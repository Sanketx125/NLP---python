#### 1. Hierarchy of Text:
   * checking no. of text (tokenize) & alphabates (n-gram) in data
   * also we check frequency of any word in data


#### 2. Tokens:
   * separating the words from data which help us to convert words into number format.
   * it has two types:
        * sentence Tokenize
        * word Tokenize


#### 3. Vocabulary:
   * checking no. of nounce, pronouns, verbes, adjectives etc in data.
   * checking unqueness of words in data also counting of them.


#### 4. Punctuation:
   * checking No. of Special symbols i.e.-->  (. , ? ! () < > : ; '' "" ') etc.


#### 5. Part of Speech:

   * adf
   * adfa


#### 6. Root of word/ Base of a Word:
   * study, studing, studies  ------> stud


#### 7. Stop Words:
   * Removing stop words
   * eg.----->.. 'the', 'of', 'to', 'this', 'do',....

```python
import string # Sample text data 
text_data = "Hello, world! This is an example: let's remove punctuation."

# Remove punctuation 
translator = str.maketrans('', '', string.punctuation) 
cleaned_text = text_data.translate(translator)
 
print(cleaned_text)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Sample text data
text_data = "This is a simple example to demonstrate how to remove stop words from a text."

# Get the list of stop words
stop_words = set(stopwords.words('english'))

# Tokenize the text
word_tokens = word_tokenize(text_data)

# Remove stop words
filtered_text = [word for word in word_tokens if word.lower() not in stop_words]

# Join the words back into a string
cleaned_text = ' '.join(filtered_text)

print(cleaned_text)
```
